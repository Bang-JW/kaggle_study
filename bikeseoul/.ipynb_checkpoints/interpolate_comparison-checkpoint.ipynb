```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv('data/train.csv')
```

```python
#‘linear’: Ignore the index and treat the values as equally spaced.
ip_linear = train[['PM10','PM2.5']].interpolate(method = 'linear') #slinear도 같은 결과
ip_pad = train[['PM10','PM2.5']].interpolate(method = 'pad') #"nearest와 동일"

# spline method -> 상세한 내용에 대해서는 좀 더 필요(결과는 가장 비슷하게 나옴)
# 단, 초미세먼지 데이터에서 음수가 나온다는 부분은 조금 아쉽
# order 는 함수의 차수를 의미 즉 여기서는 4차 함수 
ip_spline = train[['PM10','PM2.5']].interpolate(method='spline', order = 4)
```

```python
# interpolation result
plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.title('linear interpolation')
sns.lineplot(data = ip_linear['PM10'][78:100])

plt.subplot(2,2,2)
plt.title('pad interpolation')
sns.lineplot(data = ip_pad['PM10'][78:100])

plt.subplot(2,2,3)
plt.title('spline interpolation')
sns.lineplot(data = ip_spline['PM10'][78:100])
```

```python
def Baseline_NMAE(true, pred): 
    score = np.mean(np.abs(true-pred) / true) 
    return score
```
