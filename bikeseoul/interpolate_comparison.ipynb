```python
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
```

```python
train[78:100]
```

```python
#‘linear’: Ignore the index and treat the values as equally spaced.
ip_1 = train.interpolate(method = 'linear')[78:100] #slinear도 같은 결과
ip_1
```

```python
sns.lineplot(ip_1.index, ip_1['PM10'])
```

```python
#‘pad’: Fill in NaNs using existing values.
ip_2 = train.interpolate(method = 'pad')[78:100] #"nearest와 동일"
ip_2
```

```python
sns.lineplot(ip_2.index, ip_2['PM10'])
```

```python
# spline method -> 상세한 내용에 대해서는 좀 더 필요(결과는 가장 비슷하게 나옴)
# 단, 초미세먼지 데이터에서 음수가 나온다는 부분은 조금 아쉽
# order 는 함수의 차수를 의미 즉 여기서는 4차 함수 
ip_3 = train.interpolate(method='spline', order = 4)[78:100]
ip_3
```

```python
sns.lineplot(ip_3.index, ip_3['PM10'])
```
